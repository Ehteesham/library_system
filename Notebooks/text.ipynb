```python
import cv2
from flask import Flask, Response, request
from functools import wraps

app = Flask(__name__)

# Define a username and password for authentication
USERNAME = 'ehteshaam44'
PASSWORD = 'ayeman@786'

# Function to check if the provided username and password are correct
def check_auth(username, password):
    return username == USERNAME and password == PASSWORD

# Function to deny access if authentication fails
def authenticate():
    return Response(
        'Authentication required', 401,
        {'WWW-Authenticate': 'Basic realm="Login Required"'})

# Wrapper for protected routes
def requires_auth(f):
    @wraps(f)
    def decorated(*args, **kwargs):
        auth = request.authorization
        if not auth or not check_auth(auth.username, auth.password):
            return authenticate()
        return f(*args, **kwargs)
    return decorated

def generate_frames():
    camera = cv2.VideoCapture(0)  # Use the default camera (change index if necessary)

    while True:
        success, frame = camera.read()
        if not success:
            break
        else:
            ret, buffer = cv2.imencode('.jpg', frame)
            if not ret:
                continue
            yield (b'--frame\r\n'
                   b'Content-Type: image/jpeg\r\n\r\n' + buffer.tobytes() + b'\r\n')

@app.route('/')
@requires_auth  # Add authentication to access the camera feed
def video_feed():
    return Response(generate_frames(), mimetype='multipart/x-mixed-replace; boundary=frame')

@app.teardown_appcontext
def teardown_camera(exception):
    camera.release()

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000, threaded=True)
```

```python
import random
import string

def generate_username(length=8):
    # Generate a random username of the specified length
    username = ''.join(random.choices(string.ascii_lowercase, k=length))
    return username

def generate_password(length=12):
    # Generate a random password of the specified length
    password_characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(password_characters) for _ in range(length))
    return password

if __name__ == "__main__":
    num_accounts = 5  # Number of username/password pairs to generate

    
    username = generate_username()
    password = generate_password()
    print(f"Username: {username}")
    print(f"Password: {password}")
    print()
```

```python
import tkinter as tk

def disable_button():
    button.config(state="disabled")

def enable_button():
    button.config(state="normal")

root = tk.Tk()
root.geometry("300x200")

button = tk.Button(root, text="Click Me", command=disable_button)
button.pack(pady=20)

enable_button_button = tk.Button(root, text="Enable Button", command=enable_button)
enable_button_button.pack()

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

def reset_columns():
    # Clear existing columns and headings
    for column_id in tree.get_children():
        tree.delete(column_id)

    # Define and configure new columns and headings
    tree["columns"] = ("new_column1", "new_column2", "new_column3")
    tree.heading("#0", text="New Column 1")
    tree.column("#0", width=100)
    tree.heading("#1", text="New Column 2")
    tree.column("#1", width=100)
    tree.heading("#2", text="New Column 3")
    tree.column("#2", width=100)

root = tk.Tk()
root.geometry("600x400")

tree = ttk.Treeview(root)
tree["columns"] = ("column1", "column2", "column3")  # Initialize with columns
tree.heading("#0", text="Column 1")
tree.column("#0", width=100)
tree.heading("#1", text="Column 2")
tree.column("#1", width=100)
tree.heading("#2", text="Column 3")
tree.column("#2", width=100)

tree.pack()

reset_columns_button = tk.Button(root, text="Reset Columns", command=reset_columns)
reset_columns_button.pack()

root.mainloop()
```
